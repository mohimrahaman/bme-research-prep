```python
import numpy as np
```

```python
# Create a 1D array with numbers from 0 to 11
array_1d = np.arange(12)
print("Original 1D Array:\n", array_1d)

# Reshape into 3 rows and 4 columns
matrix_3x4 = array_1d.reshape(3, 4)
print("\nReshaped to 3x4 Matrix:\n", matrix_3x4)

# Reshape into 4 rows and 3 columns
matrix_4x3 = array_1d.reshape(4, 3)
print("\nReshaped to 4x3 Matrix:\n", matrix_4x3)
```

```python
# Create two 2x2 matrices
matrix_A = np.array([[1, 2], [3, 4]])
matrix_B = np.array([[5, 6], [7, 8]])

print("Matrix A:\n", matrix_A)
print("\nMatrix B:\n", matrix_B)

# Element-wise addition
add_result = matrix_A + matrix_B
print("\nElement-wise Addition (A + B):\n", add_result)

# Element-wise multiplication
mul_result = matrix_A * matrix_B
print("\nElement-wise Multiplication (A * B):\n", mul_result)
```

```python
# Create a 3x3 matrix
matrix_3x3 = np.array([[10, 20, 30],
                       [40, 50, 60],
                       [70, 80, 90]])

# Create a 1D array (vector) with 3 elements
vector = np.array([1, 2, 3])

print("3x3 Matrix:\n", matrix_3x3)
print("\n1D Vector:\n", vector)

# Broadcast addition: the vector is added to EACH row of the matrix automatically
broadcast_result = matrix_3x3 + vector
print("\nResult of Broadcasting (Matrix + Vector):\n", broadcast_result)
```

```python
import pandas as pd

# Load the Titanic dataset directly from a public URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display the first 5 rows to make sure it loaded correctly
df.head()
```

```python
# Example 1: Use .loc to get the 'Name' and 'Age' of the first 5 rows
print("--- Selecting via .loc (Labels) ---")
print(df.loc[0:4, ['Name', 'Age']])

# Example 2: Use .iloc to grab the exact same data using integer positions
# Row 0 to 5 (exclusive), and Columns 3 and 5 (Name and Age positions)
print("\n--- Selecting via .iloc (Positions) ---")
print(df.iloc[0:5, [3, 5]])
```

```python
# Filter rows where Age > 30
above_30 = df[df['Age'] > 30]

print(f"Total passengers: {len(df)}")
print(f"Passengers over 30 years old: {len(above_30)}")
print("\nFirst 3 passengers over 30:")
above_30[['Name', 'Age', 'Pclass']].head(3)
```

```python
# Group by 'Pclass' (Passenger Class: 1st, 2nd, 3rd) and calculate the average of 'Survived'
survival_by_class = df.groupby('Pclass')['Survived'].mean()

print("Survival Rate by Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd):")
print(survival_by_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean default style for all our plots
sns.set_theme(style="whitegrid")
```

```python
# Create a figure and axis with a specific size (8 inches wide, 5 inches high)
plt.figure(figsize=(8, 5))

# Plot a histogram of passenger ages using Seaborn
# 'kde=True' adds a smooth density curve over the bars
sns.histplot(data=df, x='Age', kde=True, color='teal', bins=30)

# Properly label the axes and add a title
plt.xlabel('Passenger Age', fontsize=12)
plt.ylabel('Count of Passengers', fontsize=12)
plt.title('Distribution of Passenger Ages on the Titanic', fontsize=14, fontweight='bold')

# Display the plot cleanly
plt.show()
```

```python
plt.figure(figsize=(8, 5))

# Plot a bar chart showing survival counts across different classes
# 'hue' splits the bars into color-coded categories (0 = Died, 1 = Survived)
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set2')

# Properly label axes and title
plt.xlabel('Passenger Class (1st, 2nd, 3rd)', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.title('Survival Counts Grouped by Passenger Class', fontsize=14, fontweight='bold')

# Customize the legend labels so they are easy to read
plt.legend(title='Outcome', labels=['Died', 'Survived'])

# Display the plot
plt.show()
```
